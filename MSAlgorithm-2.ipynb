What is the Merge Sort Algorithm?

It is a divide and conquer algorithm to solve and unorgainzed array into a sequence, ordered array

It splits into 2 halves, sorts them, and merges the sorted halves. 

Ex. 

43 2 24 65 34 1 67 35

Splits [Step 1]
 
(1) 43 2 24 65 

(2) 34 1 67 35

Divide by 2 [Step 2]

43 2

24 65

34 1

67 35

Organize pairs [Step 3]

(By lowest to highest)

2, 43

24, 65

1, 34

35, 67

Classify Pairs [Step 4]

2, 43 a

24, 65 b

1, 34 c 

35, 67 d

Merge a & b and c & d [Step 5]

ab 2, 24, 43, 65
cd 1, 34, 35, 67

Merge ab & cd

(Do this by comparing ab[0] to cd[0], etc...

1, 2, 24, 34, 35, 43, 65, 67

```python
#Python code
array1 = [56, 39, 66, 8, 37, 57, 15, 30, 27, 77]
```

```python
#Step 1 -- Split into 2
def ms_step1():
    middle = len(array1) / 2
    middle = int(middle)
    group1a = array1[:middle]
    group2a = array1[middle:]
    
ms_step1()
```

```python
#Step 2 -- Group
def ms_step2():
    if len(group1a) % 2 == 1:
     #Odd
        group1b = group1a[:3]
        group2b = group1a[3:]
        group3b = group2a[:3]
        group4b = group2a[3:]
        
    
    else:
        #Even
        group1b = group1a[:2]
        group2b = group1a[2:]
        group3b = group2a[:2]
        group4b = group2a[2:]
        
ms_step2()
```

```python
#Step 3 -- Organize Pairs
#Step 4 -- Classify Pairs
#Use sort function
def ms_step3a4():
    group1b.sort()
    group2b.sort()
    group3b.sort()
    group4b.sort()
    
ms_step3a4()
```

```python
# Step 5 -- Merge Lists by comparing
def ms_step5():
    bin1 = group1b + group2b
    bin2 = group3b + group4b
    bin1.sort()
    bin2.sort()
    solution = bin1 + bin2
    solution.sort()
    print(solution)
    
ms_step5()
           
```

```python
array1 = [56, 39, 66, 8, 37, 57, 15, 30, 27, 77]; print(array1)
ms_step1()
ms_step2()
ms_step3a4()
ms_step5()
#Code written by Will Zou
```

```python
def merge_sort():
    array1 = [56, 39, 66, 8, 37, 57, 15, 30, 27, 77]
    
    #step 1
    middle = len(array1) / 2
    middle = int(middle)
    group1a = array1[:middle]
    group2a = array1[middle:]
    #step 2
    if len(group1a) % 2 == 1:
     #Odd
        group1b = group1a[:3]
        group2b = group1a[3:]
        group3b = group2a[:3]
        group4b = group2a[3:]
        
    
    else:
        #Even
        group1b = group1a[:2]
        group2b = group1a[2:]
        group3b = group2a[:2]
        group4b = group2a[2:]
        
    #step 3 and 4
    group1b.sort()
    group2b.sort()
    group3b.sort()
    group4b.sort()
    
    #step 5
    bin1 = group1b + group2b
    bin2 = group3b + group4b
    bin1.sort()
    bin2.sort()
    solution = bin1 + bin2
    solution.sort()
    print(array1)
    print(solution)

merge_sort()
    
```
